```python
import arcpy
import json
import os
```

```python
basePath = 'C:\\Users\\rjc.EAGLE\\Documents\\GitHub\\AgricultureDataModel'

data = {}
data.update(json.load(open('{0}\\json\\{1}.json'.format(basePath, 'datamodel'))))
data
```

```python
oldDataPath = 'C:\\Users\\rjc.EAGLE\\Documents\\ArcGIS\\Projects\\Farm-Data-Model\\DONOTDELETE'

for service in data['services']:
    
    oldGDB = '{0}\\{1}.gdb'.format(oldDataPath,service['name'])
        
    arcpy.env.workspace = oldGDB
    oldFCs = arcpy.ListFeatureClasses()
    print(oldFCs)
    
    for fc in oldFCs:
        fcPath = '{0}\\{1}'.format(oldGDB,fc)
        print(fcPath)
    
        newfcPath = '{0}\\output\\data\{1}.gdb\\{2}'.format(basePath, service['name'], fc)
        arcpy.Append_management(fcPath, newfcPath, 'NO_TEST') 
```
